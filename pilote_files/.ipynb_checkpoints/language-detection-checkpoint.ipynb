```python
import spacy
from spacy_langdetect import LanguageDetector
import fr_core_news_md
nlp = fr_core_news_md.load()
nlp.add_pipe(LanguageDetector(), name='language_detector', last=True)
text = 'This is an english text. Mais on parle aussi autre langues. Errare humanum est, perseverare autem diabolicum.'
doc = nlp(text)
# document level language detection. Think of it like average language of the document!
print(doc._.language)

# sentence level language detection
for sent in doc.sents:
   print(sent, sent._.language)
```

```python
frenchSents = [print(sent) for sent in doc.sents if 'sen._.language' == 'fr']
```

```python
import requests 
voltaire_candide = requests.get('https://raw.githubusercontent.com/MiMoText/roman-dixhuit/master/plain/files/Voltaire_Candide.txt')
voltaire_candide = nlp(voltaire_candide.text)
for sent in voltaire_candide .sents:
   print(sent, sent._.language)
```
